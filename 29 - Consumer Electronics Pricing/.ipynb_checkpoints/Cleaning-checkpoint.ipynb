```python
import pandas as pd
import numpy as np
from plotnine import *
from mizani.formatters import comma_format, percent_format
from skimpy import clean_columns, skim
import os
import itertools
from difflib import SequenceMatcher

pd.options.display.max_columns = 500

%load_ext blackcellmagic
```

### Cleaning the Raw Data

Let's see what the results of the last week are:

```python
os.listdir("Raw Data")
```

We have three files, which we can read in:

```python
data = []

for df in os.listdir("Raw Data"):
    tmp = pd.read_csv(f"Raw Data/{df}")
    data.append(tmp)
    del tmp
```

### Device Specifications:

```python
data[0]
```

```python
data[0].duplicated(["Header", "product_id"]).sum()
```

Apparently there are a few duplicated values for the device spec and ID, let's drop those and then pivot wider, so we can actually use the data.

```python
specs = (
    data[0]
    .drop_duplicates(["Header", "product_id"])
    .pivot(index="product_id", columns="Header", values="Content")
    .reset_index()
    .rename_axis(None, axis=1)
    .pipe(clean_columns)
)

specs.head()
```

Let's check if there are some truly useless columns:

```python
(
    ggplot(
        data=specs.isna().mean().reset_index(name="missing"),
        mapping=aes(x = "missing", y = "reorder(index, missing)"),
    )
    + geom_point()
    + labs(title = "Missing values by column", x = "Missingness", y = "Column Name")
    + geom_vline(xintercept=0.8)
    + scale_x_continuous(labels=percent_format())
    + theme_light()
    + theme(figure_size=[8, 20])
)
```

Let's drop everything above 50% missingness:

```python
specs = specs.dropna(thresh=int(specs.shape[0]*(1-0.5)), axis=1)
```

From the plot above I saw some columns that had similar names, let's calculate the similarity between column headers and see which ones are repeating. 

```python
(
    pd.DataFrame(
        itertools.combinations(specs.columns, r=2), columns=["col1", "col2"]
    )
    .assign(
        similarity=lambda x: x.apply(
            lambda r: SequenceMatcher(None, r.col1, r.col2).ratio(), axis=1
        )
    )
    .sort_values("similarity", ascending=False)
    .query("similarity < 1 and similarity > 0.7")
)
```

To me, only the first case seems problematic, so let's fix that:

```python
specs = (
    specs.assign(
        pixelauflosung=lambda x: np.where(
            x.pixelauflosung.isna(), x.pixelauflosungi, x.pixelauflosung
        ),
        kamera_anzahl=lambda x: np.where(
            x.kamera_anzahl.isna(), x.kamera_anzahli, x.kamera_anzahl
        )
    ).drop(["pixelauflosungi", "kamera_anzahli"], axis=1)
)
```

```python
specs.shape
```

So now we're down to 55 columns and fixed the obvious mistakes. Let's translate the column names:

```python
specs = specs.rename(columns={
    'anzahl_prozessorkernei': 'anzahl_prozessorkerne',
    'bildwiederholrate_bei_max_auflosungi': 'bildwiederholrate_bei_max_auflosung',
    'hauptkamera_lichtstarkei': 'hauptkamera_lichtstarke',
})
```

```python
specs.to_csv("Cleaned Data/specs.csv", index=False)
```

### Prices

VAT is not really interesting, as it's constant for all, so we'll ignore it here:

```python
preise = (
    data[1]
    .pipe(clean_columns)
    .drop("amount_excl", axis=1)
    .rename(
        columns={
            "amount_incl": "preis",
            "valid_from": "datum",
        }
    )
    .assign(datum=lambda x: pd.to_datetime(x.datum.apply(lambda r: r[0:10])))
    .drop_duplicates()
)

preise
```

```python
preise.to_csv("Cleaned Data/preise.csv", index=False)
```

### Price Development Post-Launch

```python
preise = pd.read_csv("Cleaned Data/preise.csv", parse_dates=["datum"])
specs = pd.read_csv("Cleaned Data/specs.csv")
```

```python
specs.value_counts("hersteller")
```

```python
competitor_product_ids = (
    specs.query("hersteller in ['Apple', 'Samsung']")["product_id"]
    .drop_duplicates()
    .tolist()
)

plot_data = (
    preise.query("product_id in @competitor_product_ids")
    .merge(
        specs.filter(
            [
                "product_id",
                "hersteller",
                "farbgruppe",
                "modellreihe",
                "speicherkapazitat",
                "titel",
            ]
        ),
        how="left",
        on="product_id",
    )
    .sort_values(["titel", "datum"])
    .reset_index(drop=True)
    .assign(
        preis=lambda x: x.preis / x.groupby(["product_id"])["preis"].transform("first"),
        day=lambda x: x.groupby("product_id").cumcount() + 1,
    )
    .groupby(["hersteller", "day"])
    .apply(
        lambda x: pd.Series(
            {
                "lower": x.preis.quantile(0.25),
                "middle": x.preis.median(),
                "higher": x.preis.quantile(0.75),
            }
        )
    )
    .reset_index()
    .assign(
        lower = lambda x: x.groupby("hersteller")["lower"].transform(lambda y: y.rolling(14).mean()),
        middle = lambda x: x.groupby("hersteller")["middle"].transform(lambda y: y.rolling(14).mean()),
        higher = lambda x: x.groupby("hersteller")["higher"].transform(lambda y: y.rolling(14).mean()),
    )
)
```

```python
plot_data
```

```python
(
    ggplot(data=plot_data, mapping=aes(x="day"))
    + geom_line(aes(y="middle", colour="hersteller"))
    + geom_ribbon(aes(ymin="lower", ymax="higher", fill="hersteller"), alpha=0.25)
)
```
