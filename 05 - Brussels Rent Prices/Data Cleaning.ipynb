Problem with missing columns:

```python
pd.DataFrame(
    (data
     .isna()
     .sum()
     .sort_values(ascending=False)
     / len(data)
     ),
     columns=["missing"]
).head(20)
```

```python
data = data.dropna(axis=1, thresh=0.6 * data.shape[0])
```
